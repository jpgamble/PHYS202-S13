Electric Dipole Potential
-
\begin{equation}
\ V(x,y) = \fract{\kq}{\sqrt{x^2 + (y-d)^2} - \fract{\kq}{\sqrt{x^2+(y+d)^2}
\end{equation}

```python
dx  = 0.06
dy = 0.06
x = arange(-2., 2., dx)
y = arange(-2.,2., dy)
print x.ndim
print len(x)
X, Y =  meshgrid(x,y)
X.ndim
print Y.shape
print Y
```

```python
k = 1
q = 1
d = 1
V = (k*q)/((X**2 + (Y-d)**2)**(1/2.)) - (k*q)/((X**2 + (Y+d)**2)**(1/2.))
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X,Y,V,rstride = 1, cstride = 1, linewidth = 0, cmap = cm.coolwarm)
#help(Axes3D.plot_surface)
cb = fig.colorbar(p)
```

contours
-

```python
levels = array([8, 4,2,1,.5,.25, 0, -.25, -.5, -1.,-2,-4,-8],)
contour(X,Y,V,levels, linewidths=4, cmap = cm.coolwarm)
```

Ex1. Make a 3-D plot of the function
-
$G(x,y) = e^{-x^2-y^2}$
over the range -6< $(x,y)$<6 using 51 bins in $x$ and 51 bins in $y$
-

```python
dx  = 12/51.
dy = 12/51.
x = arange(-6., 6., dx)
y = arange(-6.,6., dy)
X, Y =  meshgrid(x,y)
G = exp(-(X**2) - (Y**2))
print G
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X,Y,G,rstride = 1, cstride = 1, linewidth = 0, cmap = cm.coolwarm)
#help(Axes3D.plot_surface)
cb = fig.colorbar(p)
```
