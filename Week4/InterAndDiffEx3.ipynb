```python
def Temp(Resistance):
    """This funstion allows you to take a given resistance and calculate the Temperature"""
    T = arange(10, 39, 1)
    R = array([3.239, 3.118, 3.004,2.897, 2.795, 2.7, 2.610, 2.526,2.446, 2.371, 2.3, 2.233, 2.168, 2.11, 2.053, 2, 1.95, 1.902, 1.857, 1.815, 1.77, 1.736, 1.7, 1.66, 1.634, 1.603, 1.574, 1.547, 1.521, 1.496])
    m  = (T[-1] - T[0])/(R[-1] - R[0])
    b  = (R[0])-(T[0]*m)
    temp = m*Resistance + b
    return temp
Temp(1.76)
```

```python
plot(arange(-10.0, 10.0, 1.0), Temp(arange(-10.0, 10.0, 1.0)), 'r-')
print Temp(arange(-10.0, 10.0, 1.0))
xlabel('Resistance(MegaOhms)')
ylabel('Temperture(C)')
annotate('1.7 MOhms', xy=(.7, .4),  xycoords='axes fraction',
                horizontalalignment='center', verticalalignment='center')
show()
print str(Temp(1.7)) + 'MOhms'
```

```python
def finiteDifference(x,y):
    """This function takes one array as an arguments(x), and returns the derivative of the second input (y) with repect to x."""
    dydx = diff(y)/(diff(x))
    return dydx
```

```python
x = arange(0.,10.,.1)
y = x**5
c = finiteDifference(x, y)
plot(c)
show()
```

```python
dydx = 5*x**4
dydx1 = zeros(y.shape, float)
dydx1[:-1] = diff(y)/diff(x)
dydx1[-1]  = (y[-1] - y[-2])/(x[-1]-x[-2])
plot(x, dydx, x, dydx1, 'g+')
legend(loc = 'upper left')
show()
```

```python
percentError = 100*abs(dydx1[1:] - dydx[1:])/dydx[1:]
print percentError
```

```python
x = arange(0.,10.,.001)
y = x**5
c = finiteDifference(x, y)
percentError = 100*abs(dydx1[1:] - dydx[1:])/dydx[1:]
```

```python
plot(percentError)
xlabel('X')
ylabel('Percent Error with exact dydx')
show()
```

```python
import scipy
import numpy
mask = (percentError < 1e4)  
plot(x, percentError, x[mask], percentError(x[mask]))
```

```python
def PtFiniteDiff(x,y):
    dydx= zeros(y.shape, float)
    dydx[0] = (y[1] - y[0])/(x[0]-x[1])
    dydx[-1]  = (y[-1] - y[-2])/(x[-1]-x[-2])
    dydx[1] = (y[1] - y[0])/(x[1] - x[0])
    dydx[-2] = (y[0] - y[-2] )/(x[0]- x[-2])
    for n in (y[-2:2]):
        dydx[n] = (y[n-2] - (8*y[n-1]) - (8*y[n-1])- y[n+2])/(12*(x[n+1] - x[n]))
    return dydx
```
