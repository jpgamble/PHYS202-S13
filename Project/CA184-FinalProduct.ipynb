Rule CA184  For One-Dimensional Traffic Flow 
=
The following shows an attempt at programming the STCA Ruleset or "Stochastic Trac Cellular Automata" with an addtional case for cruise control
-
The rules are as follows:
    
1. Let $g_i$ equal the number of empty sites in front of each car.

2. If $g_i$ = 0, that car does not move during this time step.

3. If $v_i$ > $g_i$, slow down to $v_i = g_i$

4. If $v_i$ < $g_i$ and $v_i < v_{max}$, speed up by one unit: $v_i = v_i + 1$.

5. For $v_i$ > 0, allow for random slowdowns by letting $v_i = v_i - 1$ with probability $p$

6. Move all cars forward by $v_i$

Rule CA184:if  $v_i = v_{max}$ and $v_i < g_i$ then $v_i = v_i$
.

```python
import random #This allows for the definition of a random element to be added to the ruleset
```

Below are three definitions of functions. 
-

In these functions we allow the movement of the traffic elements $(newpos)$ and a definition of the initial conditions $(InitialCond)$ for the ruleset ($STCA$ and $CA184$) to take place.

```python
def newpos(pos, vel, index):
    """Takes the parameters of position and velocity and pushes the new values to the array them for their movement based on their individual index."""
    pos[index] = pos[index] + vel[index] # moves each element 
    if pos[index] > road: #moves the element according to the offset of the road
        pos[index] = pos[index] - 40
    return pos
```

```python
def InitialCond(NCars, road):
    """Takes Number of Cars (NCARS) and Maximum Road Length as parameters, returning a position and velocity array for each element"""
    pos = zeros(NCars)
    vel = zeros(NCars)
    for i in range(len(pos)-1):
        pos[i] = randint(0, road)#fills the position array with values between 0 and the maximum road length
    for k in range(len(vel)-1):
        vel[k] = randint(0, road)# Fills velecoty array with values between 0 and max number of availible spaces
    return pos, vel
```

```python
def CA184(pos, vel, road, cars):
    """Implements the ruleset on the given parameters of Position and Velocity array, road length and number of cars"""
    for i in range(len(pos)-1):
        pos[i] = randint(0, road)#pushes new values into the position array
    for k in range(len(vel)-1):
        vel[k] = randint(0, road-cars)#pushes the new values into the velcoty array
    prob = rand(1) #sets the probablity of slowdown
    index = randint(0, len(pos)-1)
    carpos = pos[index]#choses a random car
    sortedpos = sort(pos)#sorts the positions to find the space between the chosen car and its surrounding constitutants 
    if carpos == (len(pos)-1): #this states that if the element is in the last possible position it space is equal to the space before the first element in the array
        gi = sortedpos[0]
        if gi == 0:
            vel[index] = 0
    else:
        #Rule 1
        gi = (sortedpos[index+1] - sortedpos[index])
        #Rule 2
        if gi == 0:
            vel[index] = 0
    #Rule 3
    if vel[index] > gi:
        vel[index] == gi
    #Rule 4
    elif vel[index] < gi and vel[index] < max(vel):
        vel[index] = vel[index] + 1
    #CA184
    elif vel[index] < gi and vel[index] == max(vel):
        vel[index] = vel[index]
    #Rule 5
    else:
        vel[index] = vel[index]*prob
        for e in range(cars):
            pos[e] = pos[e]+vel[e] # Adjusts the All positions according to velocity
    
    print "CA184 Position is: " + str(pos[index])
    print "CA184 Velocity is: " + str(vel[index])
    print "CA184 Space between car and car before: " + str(gi)
    return pos[index], vel[index], gi
```

```python
N = 150 #Sets number of time steps we are going to look at
Ncars = 10#Sets the number of cars
road = 40#sets the road length
posi1 = zeros(N)
veli1 = zeros(N)
g1 = zeros(N)
dens1 = zeros(N)
avgveli1 = zeros(N)
curr1 = zeros(N)
conds = InitialCond(Ncars, road)#sets the initial conditions for analysis
pos1= conds[0] 
#because InitialCond returns a tuple the position array will be filled with the first coloumn of values 
#and velocity array will be filled with the second coloumn of elements 
vel1 =conds[1]
for i in range(N-1):
    posi1[i], veli1[i], g1[i] = CA184(pos1, vel1, road, Ncars)
    dens1[i] = Ncars/(posi1[i])#the density is defined as Number of cars per position
    avgveli1[i] = average(veli1[i])
    curr1[i] = dens1[i]*avgveli1[i]
```
