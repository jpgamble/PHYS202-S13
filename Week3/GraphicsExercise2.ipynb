```python
import numpy
import math

i = 1j
x = arange(-pi, pi, .005*pi)
y = arange(-1, 1, .5)
U  = -(1/cosh(x))**2
a = tanh(x)
d = ((.5*i + tanh(x))/(.5*i +1))*e**(+.5*i*x)
b =  (i + tanh(x))/(i +1)*e**(+i*x)
c = (2*i + tanh(x))/(2*i +1)*e**(+2*i*x)
h = a.conj()*a
e = d.conj() * d
f = b.conj() * b
g = c.conj() * c
line = plot(x, U)
plot(x, h, 'r-')
plot(x, e, 'r--')
plot(x, f, 'r-.')
plot(x, g, 'r:')
legend(('$U(x)$', '$\psi_k^2, k = 0.0$', '$\psi_k^2, k = 0.5$', '$\psi_k^2, k = 1.0$', '$\psi_k^2, k = 2.0$'),prop = {'size' : 8}, loc='lower left', shadow=True)
grid(True)
xlabel('$x(rad)$')
ylabel('$U(x),\psi_k(x)^2$')
xticks(arange(-pi , pi+.01, pi/2), ('$-\pi$', '$-\pi/2$', '0', '$+\pi/2$', '$+\pi$'))
yticks(arange(-1, 1+.01, .5))
show()
```
