```python
x = arange(0., pi + 0.001, 0.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0) - cos(x))/2.

plot(x, f(x), label = 'True distribution', lw = 3)
plot (x, F(x), label = 'cummulative function', lw =3)

xlim(0, 3.5)
ylim(0,1.1)

legend(loc = 0)
show()
```

Exercise 1: Expoential function - analytic
    

```python
x = arange(0., 5 + 0.001, 0.01)
L = 1
f = lambda x: L*e**(-L*x)
F = lambda x: -e**(-L*x)
Finv = lambda x : -log(x)/L
plot(x, f(x), label = 'True distribution', lw = 3)
plot (x, F(x), label = 'cummulative function', lw =3)
plot (x, Finv(x), label = 'Inverse function', lw =3)
xlim(0, 3.5)
ylim(-1.1,1.1)

legend(loc = 0)
show()
```

```python
N = 10000

z = numpy.random.random_sample(N);

Finv = lambda z : -log(z)/L
xsamples = Finv(z);
hist(xsamples, 50, normed=True, label = "Sampled Distribution")
plot(x, f(x), label = 'True distribution', lw = 3, color = 'red')

xlim(0, 5)
ylim(-2.01, 2)

legend(loc = 0)
show()
```
