Curve fitting example

```python
from scipy.optimize import curve_fit
```

```python
def funct(x, a, b, c): 
    return (a*x**2 + b*x + c)
```

```python
x = array([0.0, 1.0, 2., 3, 4, 5])
y = array([-1.78, 4.09, 8.85, 17.9, 26.1, 35.2])
yerr = array([0.46, 0.658, .528, 1.34, 1.09, 0.78])

p0 = [1., 3., -2]
popt, pcov = curve_fit(funct, x, y, p0, yerr)
```

```python
punc= zeros(len(popt))
for i in arange(0, len(popt)):
    punc[i] = sqrt(pcov[i, i])
print "optimal parameters: ", popt
print "uncertainties of parameters: ", punc
```

```python
errorbar(x,y,yerr, fmt = 'r*')
fitresult = funct(x, popt[0], popt[1], popt[2])
plot(x, fitresult, 'b-')
show()
```

curve fitting example 2

```python
waveform_1 = load('waveform_1.npy')
t = np.arange(len(waveform_1))
```

```python
plot(t, waveform_1)
xlabel("Time(ns)")
ylabel("Intensity (bin)")
show()
```

```python
def model(t, B, A, mu, sigma):
    return B + A*exp(-((t-mu)/sigma)**2)
```

```python
x0 = array([3., 30, 15, 1])
```

```python
popt, pcov = curve_fit(model, t, waveform_1, x0)
plot(t, waveform_1, label = "data")
plot(t, model(t, popt[0], popt[1],popt[2] , popt[3]), label = "fit", linewidth = 2)
legend(loc = "best")
show()
```

```python
waveform_2 = load('waveform_2.npy')
t = np.arange(len(waveform_2))
plot(t, waveform_2)
xlabel("Time(ns)")
ylabel("Intensity (bin)")
show()
```

```python
def model2(t, B, A, mu, sigma, C, D, mu1, mu2, sigma1, sigma2):
    return B + A*exp(-((t-mu)/sigma)**2) + C*exp(-((t-mu1)/sigma1)**2) + D*exp(-((t-mu2)/sigma2)**2)
```

```python
x0 = array([3., 30, 15, 1, 12, 13, 10, 10, 10, 3])
popt, pcov = curve_fit(model2, t, waveform_2, x0)
plot(t, waveform_2, label = "data")
plot(t, model2(t, popt[0], popt[1],popt[2], popt[3], popt[4],popt[5], popt[6], popt[7], popt[8], popt[9]), label = "fit", linewidth = 2)
legend(loc = "best")
show()
```
