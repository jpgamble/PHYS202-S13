```python
def PtFiniteDiff(x,y):
    dydx= zeros(y.shape, float)
    dydx[0] = (y[1] - y[0])/(x[0]-x[1])
    dydx[-1]  = (y[-1] - y[-2])/(x[-1]-x[-2])
    dydx[1] = (y[1] - y[0])/(x[1] - x[0])
    dydx[-2] = (y[0] - y[-2] )/(x[0]- x[-2])
    for n in (y[-2:2]):
        dydx[n] = (y[n-2] - (8*y[n-1]) - (8*y[n-1])- y[n+2])/(12*(x[n+1] - x[n]))
    return dydx
```

```python
x = linspace(0,2*pi,100)
y = sin(x) + 0.1*random.random(size=x.shape)
j = PtFiniteDiff(x,y)
dydx = zeros(y.shape, float)
dydx[:-1] = diff(y)/diff(x)
dydx[-1]  = (y[-1] - y[-2])/(x[-1]-x[-2])
plot(x,y, label = 'original')
plot(x, j, label = '4pt Diff')
plot(x, dydx, label = '2ptDiff')
legend(loc = 'upper right')
show()
```

```python
xmin= 1.7
xmax = 12.4

subx = x[logical_and((x>xmin), (x<xmax))]
ymax = max(f(subx))*1.05
ymin = 0

myplot = plot(x, f(x), lw=3)
xlim(0., 15.)

plot([xmin, xmin], [ymin, ymax], color = 'k', linestyle = '--')
plot([xmax, xmax], [ymin, ymax], color = 'k', linestyle = '--')
plot([xmin, xmax], [ymax, ymax], color = 'k', linestyle = '--')
```

```python
N = 10000
samples_x = xmin + (xmax - xmin)*random_sample(N);
samples_y  = ymin + (ymax-ymin)*random_sample(N);

myplot = plot(x, f(x), lw=3)
xlim(0., 15.)

plot(samples_x[1:200], samples_y[1:200], '.')

plot([xmin, xmin], [ymin, ymax], color = 'k', linestyle = '--')
plot([xmax, xmax], [ymin, ymax], color = 'k', linestyle = '--')
plot([xmin, xmax], [ymax, ymax], color = 'k', linestyle = '--')
show()
```

```python
newmask = (samples_y < f(samples_x))
sum(newmask)
```

```python
myplot = plot(x, f(x), lw= 3)
xlim(0., 15.)
plot(samples_x[1:200], samples_y[1:200], '.')
plot(samples_x[newmask[:200]], samples_y[newmask[:200]], '.')

plot([xmin, xmin], [ymin, ymax], color = 'k', linestyle = '--')
plot([xmax, xmax], [ymin, ymax], color = 'k', linestyle = '--')
plot([xmin, xmax], [ymax, ymax], color = 'k', linestyle = '--')
show()
```
